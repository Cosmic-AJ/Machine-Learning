## Hyper parameter Tuning

### Number of Neurons in hidden layer

```python
neu_num = []
neu_score = []
for i in range(10,151,10):
    mlp = MLPClassifier(hidden_layer_sizes=(i,), random_state=42)
    mlp.fit(X_train_new, Y_train)
    neu_score.append(mlp.score(X_test_new, Y_test))
    neu_num.append(i)
    
plt.figure(figsize=(6, 5))
plt.plot(neu_num, neu_score)
plt.xlabel('Number of Neurons in hidden layer')
plt.ylabel('Accuracy')
plt.show()
```

```python
max(neu_score)
```

```python
neu_score[13]
```

```python
neu_num[13]
```

### Activation Function

```python
func = ['identity', 'logistic', 'tanh', 'relu']
ac_score = []
for i in func:
    mlp = MLPClassifier(hidden_layer_sizes=(140,), activation=i, random_state=42)
    mlp.fit(X_train_new, Y_train)
    ac_score.append(mlp.score(X_test_new, Y_test))
    
plt.figure(figsize=(6, 5))
plt.bar(func, ac_score)
plt.xlabel('Activation Function')
plt.ylabel('Accuracy')
plt.show()
```

```python
max(ac_score)
```

```python
ac_score[3]
```

### Number of layers

#### With 2 hidden layers

```python
lay2_num = []
lay2_score = []
for i in range(10,101,10):
    mlp = MLPClassifier(hidden_layer_sizes=(140,i), activation='relu', random_state=42)
    mlp.fit(X_train_new, Y_train)
    lay2_score.append(mlp.score(X_test_new, Y_test))
    lay2_num.append(i)
    
plt.figure(figsize=(6, 5))
plt.plot(lay2_num, lay2_score)
plt.xlabel('Number of Neurons in 2 hidden layer')
plt.ylabel('Accuracy')
plt.show()
```

```python
max(lay2_score)
```

```python
lay2_score[7]
```

```python
lay2_num[7]
```

#### With 3 hidden layers

```python
lay3_num = []
lay3_score = []
for i in range(10,101,10):
    mlp = MLPClassifier(hidden_layer_sizes=(140,80,i), activation='relu', random_state=42)
    mlp.fit(X_train_new, Y_train)
    lay3_score.append(mlp.score(X_test_new, Y_test))
    lay3_num.append(i)
    
plt.figure(figsize=(6, 5))
plt.plot(lay3_num, lay3_score)
plt.xlabel('Number of Neurons in 3 hidden layer')
plt.ylabel('Accuracy')
plt.show()
```

```python
max(lay3_score)
```

### Solver

```python
so = ['lbfgs', 'sgd', 'adam']
so_score = []
for i in so:
    mlp = MLPClassifier(hidden_layer_sizes=(140,80), activation='relu', solver=i,
                        random_state=42)
    mlp.fit(X_train_new, Y_train)
    so_score.append(mlp.score(X_test_new, Y_test))
    
plt.figure(figsize=(6, 5))
plt.bar(so, so_score)
plt.xlabel('Solver')
plt.ylabel('Accuracy')
plt.show()
```

```python
max(so_score)
```

```python
so_score[2]
```

### Learning Rate

```python
lr = [0.000001, 0.0001, 0.0005, 0.001]
lr_score = []
for i in lr:
    mlp = MLPClassifier(hidden_layer_sizes=(140,80), activation='relu', solver='adam', learning_rate_init=i,
                        random_state=42)
    mlp.fit(X_train_new, Y_train)
    lr_score.append(mlp.score(X_test_new, Y_test))
    
plt.figure(figsize=(6, 5))
plt.plot(lr, lr_score)
plt.xlabel('Learning rate')
plt.ylabel('Accuracy')
plt.show()
```

```python
max(lr_score)
```

```python
lr_score[3]
```

### Alpha Value

```python
alpha = [0.000001, 0.0001, 0.0005, 0.001]
al_score = []
for i in alpha:
    mlp = MLPClassifier(hidden_layer_sizes=(140,80), activation='relu', solver='adam', learning_rate_init=0.001,
                        alpha=i, random_state=42)
    mlp.fit(X_train_new, Y_train)
    al_score.append(mlp.score(X_test_new, Y_test))
    
plt.figure(figsize=(6, 5))
plt.plot(alpha, al_score)
plt.xlabel('Alpha Value')
plt.ylabel('Accuracy')
plt.show()
```

```python
max(al_score)
```

```python
al_score[3]
```

```python
alpha[3]
```

## Final model

```python
mlp = MLPClassifier(hidden_layer_sizes=(140,80), activation='relu', solver='adam', learning_rate_init=0.001,
                        alpha=0.001, random_state=42)
mlp.fit(X_train_new, Y_train)
mlp.score(X_test_new, Y_test)
```
