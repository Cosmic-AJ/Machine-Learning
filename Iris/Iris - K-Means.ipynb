# K-means with Iris Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings(action='ignore')
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
iris = load_iris(as_frame=True)
```

```python
data = pd.DataFrame(iris.data)
```

```python
data['target'] = iris.target
```

```python
data.head()
```

```python
data.tail()
```

```python
sns.pairplot(data)
```

```python
cor = data.corr()
plt.figure(figsize=(8, 8))
sns.heatmap(cor, xticklabels=cor.columns, yticklabels=cor.columns, annot=True)
```

## Kmeans with k value 3

```python
kmeans = KMeans(3)
kmeans.fit(data.drop(columns=['target']))
```

```python
predicted = kmeans.fit_predict(data.drop(columns=['target']))
predicted
```

```python
plt.scatter(data['sepal length (cm)'],data['target'],cmap='rainbow')
```

```python
new_Data = data.drop(columns=['target']).copy()
new_Data['Clusters'] = predicted
plt.scatter(new_Data['sepal length (cm)'],new_Data['Clusters'],c=new_Data['Clusters'],cmap='rainbow')
```

```python
plt.scatter(new_Data['sepal width (cm)'],new_Data['Clusters'],c=new_Data['Clusters'],cmap='rainbow')
```

```python
plt.scatter(new_Data['petal length (cm)'],new_Data['Clusters'],c=new_Data['Clusters'],cmap='rainbow')
```

```python
plt.scatter(new_Data['petal width (cm)'],new_Data['Clusters'],c=new_Data['Clusters'],cmap='rainbow')
```

```python
wcss=[]
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(data.drop(columns=['target']))
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,10)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
