### Splitting the dataset
WE split the dataset into 75:25 for training and testing.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data.drop(columns=['Category']),pd.DataFrame(data.Category), random_state=42, test_size=0.25)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

### Scaling the data
We scale the data to normalize all the values in all the columns

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Minimum Sample Leaf
We can select 1 minimum sample leaf for getting maximum accuracy

```python
min_leaf_val = []
min_leaf_acc = []

for i in range(1,11):
    model = DecisionTreeClassifier(criterion='gini', splitter='best', random_state=42, max_features=5,
                                   max_depth=9, min_samples_leaf=i)
    model.fit(X_train_new, Y_train)
    min_leaf_acc.append(model.score(X_test_new, Y_test))
    min_leaf_val.append(i)
    
plt.plot(min_leaf_val, min_leaf_acc)
plt.xlabel('Minimum Sample Leaf')
plt.ylabel('Accuracy')
plt.show()
```

### Tuning Min Sample Split
We can select 0.001 and min sample split for getting maximum accuracy

```python
spl_val = []
spl_acc = []

for i in range(1,10):
    j = i/1000
    model = DecisionTreeClassifier(criterion='gini', splitter='best', random_state=42, max_features=5,
                                   max_depth=9, min_samples_leaf=1, 
                                   min_samples_split=j)
    model.fit(X_train_new, Y_train)
    spl_acc.append(model.score(X_test_new, Y_test))
    spl_val.append(i)
    
plt.plot(spl_val, spl_acc)
plt.xlabel('Minimum Sample Split')
plt.ylabel('Accuracy')
plt.show()
```

```python
max(spl_acc)
```

```python
model = DecisionTreeClassifier(criterion='gini', splitter='best', random_state=42, max_features=5,
                                   max_depth=9, min_samples_leaf=1, min_samples_split=0.001)
model.fit(X_train_new, Y_train)
model.score(X_test_new, Y_test)
```

### Tuning Min Impurity Decrease
We can select 0 as min impurity decrease for getting maximum accuracy

```python
impurity_val = []
impurity_acc = []

for i in range(0,10):
    j = i/10
    model = DecisionTreeClassifier(criterion='gini', splitter='best', random_state=42, max_features=5,
                                   max_depth=9, min_samples_leaf=1, min_samples_split=0.001,
                                   min_impurity_decrease=j)
    model.fit(X_train_new, Y_train)
    impurity_acc.append(model.score(X_test_new, Y_test))
    impurity_val.append(i)
    
plt.plot(impurity_val, impurity_acc)
plt.xlabel('Minimum Impurity Decrease')
plt.ylabel('Accuracy')
plt.show()
```

### Learning Rate

```python
lr = [0.000001, 0.0001, 0.0005, 0.001]
lr_score = []
for i in lr:
    mlp = MLPClassifier(hidden_layer_sizes=(100,30), activation='relu', solver='adam', learning_rate_init=i,
                        random_state=42)
    mlp.fit(X_train_new, Y_train)
    lr_score.append(mlp.score(X_test_new, Y_test))
    
plt.figure(figsize=(6, 5))
plt.plot(lr, lr_score)
plt.xlabel('Learning rate')
plt.ylabel('Accuracy')
plt.show()
```

```python
max(lr_score)
```

### Alpha Value

```python
alpha = [0.000001, 0.0001, 0.0005, 0.001]
al_score = []
for i in alpha:
    mlp = MLPClassifier(hidden_layer_sizes=(100,30), activation='relu', solver='adam', learning_rate_init=0.001,
                        alpha=i, random_state=42)
    mlp.fit(X_train_new, Y_train)
    al_score.append(mlp.score(X_test_new, Y_test))
    
plt.figure(figsize=(6, 5))
plt.plot(alpha, al_score)
plt.xlabel('Alpha Value')
plt.ylabel('Accuracy')
plt.show()
```
