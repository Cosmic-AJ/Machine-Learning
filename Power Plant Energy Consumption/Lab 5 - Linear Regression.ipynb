## Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=42, test_size=0.25)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

## Creating the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_train_predict = model.predict(X_train)
```

```python
rmse = (np.sqrt(mean_squared_error(Y_train, Y_train_predict)))
r2 = r2_score(Y_train, Y_train_predict)
mse = mean_squared_error(Y_train, Y_train_predict)
mae = mean_absolute_error(Y_train, Y_train_predict)
```

```python
print('MSE score on Training data is {}'.format(mse))
print('MAE score on Training data is {}'.format(mae))
print('RMSE score on Training data is {}'.format(rmse))
print('R2 score on Training data is {}'.format(r2))
```

```python
Y_test_predict = model.predict(X_test)
```

```python
rmse_test = (np.sqrt(mean_squared_error(Y_test, Y_test_predict)))
r2_test = r2_score(Y_test, Y_test_predict)
mse_test = mean_squared_error(Y_test, Y_test_predict)
mae_test = mean_absolute_error(Y_test, Y_test_predict)
```

```python
print('MSE score on Test data is {}'.format(mse_test))
print('MAE score on Test data is {}'.format(mae_test))
print('RMSE score on Test data is {}'.format(rmse_test))
print('R2 score on Test data is {}'.format(r2_test))
```

```python
print('Model Intercept is {}'.format(model.intercept_))
print('Model Coefficients are {}'.format(model.coef_))
```

# Inference

We have uses Power Plant dataset for implementing Linear Regression. The dataset has 5 columns and 9568 rows. We are using this data set to find the net hourly electrical energy output of the plant by using different attributes. We have done univariate and bivariate analysis to understand the data and relationship between different attributes. We have divided the data into train and test ratio with 75:25. We have done hyperparameter tuning to tune the algorithm and we get a r2 score of 0.93. We try to use different combination of columns and at last we can conclude that by using all 4 columns together gives us the best accuracy. The rmse and r2_score is visualized using bar chart and line chart for clear understanding. Normalization doesn't gives us better rmse score in this case. Normalization helps in improving the accuracy and r2_Score for any problem. In our case, the r2_score before and after normalizing is same ensuring similar response to the given data. We have identified the coefficient and intercept value for the given problem which are (-14.78, -2.97, 0.34, -2.31) and 454.37.
