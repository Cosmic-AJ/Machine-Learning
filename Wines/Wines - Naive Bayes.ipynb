## Checking without preprocessing

### Splitting the dataset

We split the data in the ratio of 0.8:0.2 for training and testing our model.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data.drop(columns=['Class']),pd.DataFrame(data.Class), random_state=42, test_size=0.2)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

In the first approach we create the model wwithout preprocessing the values. We will save the accuracy of all the different models in a list and try to compare them using bar chart. The different methods used are Gaussian, Multinomial and Complement.

```python
from sklearn.metrics import classification_report, confusion_matrix
```

### Selecting Different Columns

We try to select different columns to check that we can reduce the number of columns for the training or not and to find out the relationship between number of columns and the accuracy. In the below selection we are selecting columns on a linear basis.

```python
features_count = []
features_accuracy = []

for i in range(1,14):
    X_train_new = X_train.iloc[:,:i]
    gnb = GaussianNB()
    gnb.fit(X_train_new, Y_train)
    
    y_pred = gnb.predict(X_test.iloc[:,:i])
    
    features_accuracy.append(gnb.score(X_test.iloc[:,:i], Y_test))
    features_count.append(i)

    
    print(str(i)+' Features Selected', '\nAccuracy: ',str(features_accuracy[i-1]),'\n')
```

```python
plt.plot(features_count, features_accuracy)
plt.xlabel('Features Selected')
plt.ylabel('Accuracy')
plt.show()
```

From the above graph, we can find the most significant columns that helps us in model classification. We try to create model using these selected features and we get an accuracy of 100%. So from this we can conclude that the 5 columns 'Alcholol', 'Ash', 'Alcanity', 'Total Phenol', 'Proline' can be used for model creating.

```python
gnb = GaussianNB()
gnb.fit(X_train.loc[:, ['Alcholol', 'Ash', 'Alcanity', 'Total Phenol', 'Proline']], Y_train)
gnb.score(X_test.loc[:, ['Alcholol', 'Ash', 'Alcanity', 'Total Phenol', 'Proline']], Y_test)
```

```python
score
```

From the below graph, we can say that we get highest accuracy with the Gaussian Method of Naive Bayes classifier.

```python
plt.figure(figsize=(6, 5))
plt.bar(['GNB', 'MNB', 'CNB'], score)
plt.xlabel('Method Selected')
plt.ylabel('Accuracy')
plt.show()
```

The dataset used is the wines dataset which provides the information that classifies the wine into three different categories. We have different features which tells that the wine is of which category. Some of the features are: Alcholol, Malic acid, Ash, Alcanity, Magnesium, etc. There are 14 columns and around 178 rows in the dataset. We first try to understand different features available in the data set. The bivariate and univariate analysis is performed to find relationship between different features and check the distributon of the data. 

- The model is first created without normalizing the values. We divide the data in the ratio of 80:20. We get an accuracy of 100% with Gaussian method of Navie Bayes and 88% and 69% with Multinomial and Complement method. The comparison is visualized using bar chart. We also try to use different number of columns for training and the accuracy increases when the number of columns selected are increased. We can select 5 columns only for getting maximum accuracy which include Alcholol, Ash, Alcanity, Total Phenol, Proline.
- The model is then normalized using the Standard Scaler and divided into Train and Test with the ratio of 0.8:0.2. We again get an accuracy of 100% with Gaussian NB and 88% and 69% with Multinomial and Complement method of Naive Bayes. The comparison is then visualized. The accuracy after selecting different columns were also same and it is visualized using Line chart. Same as above, we can select 5 columns only for getting maximum accuracy which include Alcholol, Ash, Alcanity, Total Phenol, Proline.

From the analysis, we can conclude that Normalizing doesn't improves the performance in our case. The reason could be:
- The data is already in normal distribution therefore normalizing doesn't affect the model much. It is eveident from the Univariate Analysis.
- The accuracy achieved without normalizing was 100% and we can't increase accuracy of any model which have an accuracy of 100%.

Normalizing the data helps in those problem where the data is not normally distributed or containers outlier. By using different measures and concepts, we have reduced the number of columns required for model creation.
