### map function is used to apply some function on the whole column of the data set or map the existing categorical data with some other values. 

The lower limit for any numerical data is Quartile 1 - 1.5 * Inter Quratile Range and upper limit is Quartile 3 + 1.5 * Inter Quartile Range.
The outliers of the Sale price column are removed using this technique

```python
q1 = data['Sale Price'].quantile(0.25)
q3 = data['Sale Price'].quantile(0.75)

iqr = q3-q1

lower_limit = q1-1.5*iqr
upper_limit = q3+1.5*iqr
```

```python
data['Sale Price'] = data['Sale Price'].map(lambda x: upper_limit if x>upper_limit else (lower_limit 
                                                                                         if x<lower_limit else
                                                                                        x))
```

```python
plt.boxplot(data['Sale Price'])
plt.ylabel('Sale Price')
plt.ylabel('Values')
plt.title('Finding out outliers')
plt.show()
```
